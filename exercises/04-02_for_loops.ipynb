## What's a Loop?

First, let's remember what a loop is. Imagine you're jumping on a trampoline. Each jump is like one loop. Now, what if you decide to jump exactly 10 times? That's where a for loop comes in handy!

## For Loops: The Basics

A "for loop" is a way to repeat an action a certain number of times. It's like saying, "Hey Python, do this thing these many times."

## The Syntax of a For Loop

Here's how you write a for loop in Python:


```python
for i in range(10):
    print("I'm jumping on a trampoline! "+str(i))
```


1) **`for i in range(10)`:** This line sets up a loop that will run 10 times.
   `range(10)` creates a sequence of numbers from 0 to 9. 

2) **`i`:** In each loop, `i` takes on the next number in the sequence created
   by `range(10)`. First, `i` is 0, then 1, then 2, and so on, up to 9. 

3) **`print("I'm jumping on a trampoline! " + str(i))`:** Each time the loop
   runs, this line prints out a message. The `str(i)` part converts the number
   `i` into a string so it can be added to the text. So, you'll see "I'm jumping
   on a trampoline! 0", then "I'm jumping on a trampoline! 1", and so on, up to
   "I'm jumping on a trampoline! 9". 

In summary, `i` changes with each loop, counting up from 0 to 9, and each time,
it's included in the message that's printed out. It's like having a counter that
keeps track of how many jumps you've made on the trampoline! üåüü§∏‚Äç‚ôÇÔ∏èüîÑ 

## Using For Loops with range()

The `range()` function is super handy with for loops. It's like setting the
number of jumps on a trampoline. For example, if you write `range(5)`, it's like
planning to do 5 jumps. 

The `range()` function creates a series of numbers:

- **`range(5)`**: This creates a series of numbers from 0 to 4. It's important
  to remember that the counting starts at 0 and goes up to, but does not
  include, the specified value. So, it ends at 4. 

- **`range(1, 5)`**: This creates a series of numbers from 1 to 4. Here, the
  counting starts at 1 (the first value) and goes up to, but not including, 5
  (the second value). So, it stops at 4. 

- **`range(1, 11, 2)`**: This creates a series of every other number from 1 to
  10. The counting starts at 1 (the first value), goes up to, but not including,
  11 (the second value), and skips every second number (as specified by the
  third value). So, you get 1, 3, 5, 7, 9. 

### Example Time!

Let's say you want to count from 1 to 5. Here's how you'd do it:


```python
for number in range(1, 6):
    print(number)
```


This loop will print the numbers 1, 2, 3, 4, and 5. We started at 1 and went up
to (but not including) 6.

## Challenge 4 (Bonus)

![challenge 2 image](./images/04/4.png)

Did you know you can put a `for` loop inside a `for` loop? Check it out!

Putting a `for` loop inside another `for` loop is called a "nested loop." It's like having a loop within a loop, and it's super useful for certain tasks. Let me explain a bit more:

Imagine you're organizing a small race for each grade in your school. You could use one `for` loop to go through each grade, and inside that loop, you could have another `for` loop to handle each student in that grade. It's like looping through the grades, and for each grade, looping through its students.

Here's a simple example:


```python
for grade in range(1, 4):  # Loop through grades 1 to 3
    for student in range(1, 11):  # Loop through 10 students in each grade
        print(f"Grade {grade}, Student {student}")
```


This code will print out a list for students in grades 1, 2, and 3, counting up to 10 students in each grade.

Nested loops are a powerful tool in coding, especially when you're dealing with things that are organized in rows and columns, like grids or tables! üîÑüè´üë©‚Äçüè´üë®‚Äçüè´
